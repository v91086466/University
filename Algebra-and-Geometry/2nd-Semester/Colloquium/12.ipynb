# 12. Линейное афинное многообразие в линейном пространстве. Гиперплоскость в евклидовом и унитарном пространстве.

Определения, данные для вещественного пространства, справедливы и для линейного пространства над произвольным полем $ P $.

Пусть $ V $ - линейное пространство, $ L $ - некоторое его подпространство, $ x_0 $ - некоторый вектор этого пространства. Множество $ H = \{ x_0 + x | x \in L \} $ называется линейным многообразием (или линейным афинным многообразием) пространства $ V $. Вектор $ x_0 $ - вектор сдвига, а $ L $ - направляющее подпространство.

Два линейных многообразия $ H_1 = x_1 + L_1, H_2 = x_2 + L_2 $ называются параллельными, если либо $ L_1 \subset L_2 $, либо $ L_2 \subset L_1 $.

### Теорема 67.1

Если два линеных многообразия с непустым пересечением параллельны, то одно из них содержит другое.

### Доказательство

Находим общий вектор, делаем его вектором сдвига

Profit.

### Теорема 67.2

Непустое пересечение линейных многообразий $ H_i = x_i + L_i, \space i = \overline{1, k} $ является линейным многообразием с направляющим подпространством $ \bigcap\limits_{i = 1}^k L_i $

### Доказательство

По условию $ \exists x_0 \in \bigcap\limits_{i = 1}^k H_i \Rightarrow H_i = x_0 + L_i, \space i = \overline{1, k} $. Тогда $ \bigcap\limits_{i = 1}^k H_i = x_0 + \bigcap\limits_{i = 1}^k L_i $, так как для этих множеств имеет место двустороннее вложение.

### Теорема 67.3

Всякое $ k $ - мерное линейное многообразие в $ n $ - мерном пространстве можно представить в виде пересечения $ n - k $ гиперплоскостей.

### Доказательство

Пусть $ H = x_0 + L $ - $k$-мерное линейное многообразие. Пусть $ e_1, \dots, e_k $ - базис подпространства $ L $. Дополним его до базиса пространства $ e_1, \dots, e_n $. В качестве искомых гиперплоскостей возьмём $ H_i = x_0 + L_i, i = \overline{1, n - k} $, где $ L_i $ - линейная оболочка всех векторов базиса кроме вектора $ e_{k + i} $. Очевидно, что $ L = \bigcap\limits_{i = 1}^k L_i $, а $ H = \bigcap\limits_{i = 1}^k H_i $ (в силу теоремы 67.2)

Profit.

### Теорема 67.4

Пересечение линейного многообразия $ H = x_0 + L $ с любым подпространством, дополнительным к $ L $ состоит ровно из одного вектора.

### Доказательство

Пусть $ L^\delta $ - дополнительное к $ L $ подпространство. Так как $ L \oplus L^\delta = V, x_0 = y + z, y \in L, z \in L^\delta $. Тогда $ z = x_0 - y \in L $ и это наш искомый вектор.

При доказательстве единственности опираемся на то, что $ L \cap L^\delta = \{ \theta \} $

Линейное многообразие размерности 1 в $ n $ - мерном пространстве называется прямой в линейном пространстве, а многообразие размерности $ n - 1 $ - гиперплоскостью.

Пусть $ H = x_0 + L $ - гинерплоскость в $ E(U) $. Тогда $ L^\perp $ - одномерное подпространство и его базис состоит из одного вектора $ a $.

Вектор $ x \in H \Leftrightarrow x - x_0 \in L \Leftrightarrow (x - x_0, a) = 0 $.  
Данное уравнение может быть записано в равносильной форме: $$ (x, a) = p $$ где $ p = (x_0, a) $
